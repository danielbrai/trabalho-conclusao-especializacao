```python
starter_dataset = pd.read_feather('data/processed_data')
starter_dataset
```

```python
dataset_cleaned_linebreaks = starter_dataset.copy()
dataset_cleaned_linebreaks['oracle_text'] = dataset_cleaned_linebreaks.apply(lambda row: row['oracle_text'].replace("\n", " "), axis=1)
dataset_cleaned_linebreaks
```

```python
from textblob import TextBlob
```

```python
tokens = TextBlob(dataset_cleaned_linebreaks['oracle_text'][0].lower()).words
print(tokens)
print(len(tokens))
```

### Stop Words

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
sw = stopwords.words('english')
```

```python
stopwords_cleaned = [i for i in tokens if i.lower() not in sw]
print(stopwords_cleaned)
print(len(stopwords_cleaned))
```

### Stemming

```python
from nltk.stem import PorterStemmer
```

```python
st = PorterStemmer()
```

```python
stemmed = [st.stem(w) for w in stopwords_cleaned]
print(stemmed)
print(len(stemmed))
```

### Lemmatization

```python
from textblob import Word
```

```python
lemmatized = [Word(w) for w in stopwords_cleaned]
print(lemmatized)
print(len(lemmatized))
```
